# GroundWater Water_level Prediction in India

* The Pandas library in Python provides excellent, built-in support for time series data.
* Pandas represents time series datasets as a Series.
* A Series is a one-dimensional array with a time label for each row.
* A cleaned_df is a collection of series

**DATASETS**
1. **[raw_data](#Loading-cleaned_df-from-csv-file)** ------- Raw cleaned_df
2. **[cleaned_df](#Removing-Duplicates)** ------- Data Cleaning (Duplicates removed, Null Treatment, Outlier Treatment) 
3. **[data_POMRB](#Made-a-copy-of-the-original-cleaned_df)** ------- Data with only State, District, Quarter, POMRB
4. **[data_PREMON](#Made-a-copy-of-the-original-cleaned_df)** ------- Data with only State, District, Quarter, PREMON 
5. **[data_MONSOON](#Made-a-copy-of-the-original-cleaned_df)** ------- Data with only State, District, Quarter, MONSOON
6. **[data_POMKH](#Made-a-copy-of-the-original-cleaned_df)** ------- Data with only State, District, Quarter, POMKH
7. **[df](#cleaned_df-with-waterWater_level-of-all-quarters-in-single-column)** ------- Concat data_POMRB, data_PREMON, data_MONSOON, data_POMKH

# Data Cleaning

### Removing Duplicates

```python
cleaned_df = raw_data.copy()
cleaned_df.drop_duplicates(keep='first',inplace=True,ignore_index=True)
```

```python
print(f'\n Total Duplicate Values: {len(cleaned_df[cleaned_df.duplicated(keep=False)])}\n')
```

```python
print(f'Shape of the cleaned_df: {cleaned_df.shape}')
```

### Null Treatment

```python
#heatmap to check for null values in the cleaned_df
sns.heatmap(cleaned_df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
plt.title("Missing Data Heatmap")
plt.show()
```

#### Replacing Null values with Median for POMRB, PREMON, MONSOON & POMKH
Median is used to replace missing values because it is less sensitive to outliers.


```python
cleaned_df.POMRB.fillna(cleaned_df.POMRB.quantile(0.5),inplace = True)
cleaned_df.PREMON.fillna(cleaned_df.PREMON.quantile(0.5),inplace = True)
cleaned_df.MONSOON.fillna(cleaned_df.MONSOON.quantile(0.5),inplace = True)
cleaned_df.POMKH.fillna(cleaned_df.POMKH.quantile(0.5),inplace = True)
```

```python
print('\n Null Values in each column after treating null values\n')
cleaned_df.isnull().sum()
```

### Outliers Treatment

### Post Monsoon Rabi (January to March)

```python
print('\nMean : ',cleaned_df.POMRB.mean())
print('Median : ',cleaned_df.POMRB.median())
print('Mode : ',cleaned_df.POMRB.mode()[0])
print('Max : ',cleaned_df.POMRB.max())
print('Min : ',cleaned_df.POMRB.min())
print('Unique : ',len(cleaned_df.POMRB.unique()))
print('Variance : ',cleaned_df.POMRB.var())
print('Null : ',cleaned_df.POMRB.isna().sum())
print('\n99th quantile value of Post Monsoon Rabi is : ',round(cleaned_df.POMRB.quantile(0.99),2))
print('Number of outliers above 99th quantile : ',len(cleaned_df[cleaned_df.POMRB > cleaned_df.POMRB.quantile(0.99)]))
cleaned_df.POMRB.plot(kind='box')
plt.title('Box Plot of Post Monsoon Rabi (January to March)')
```

#### Since there is no outlier above 99th quantile, no treatment is needed for this POMRB. 

#### Pre-Monsoon (April - June)

```python
cleaned_df.PREMON.plot(kind='box')
print('\nMean : ',cleaned_df.PREMON.mean())
print('Median : ',cleaned_df.PREMON.median())
print('Mode : ',cleaned_df.PREMON.mode()[0])
print('Max : ',cleaned_df.PREMON.max())
print('Min : ',cleaned_df.PREMON.min())
print('Unique : ',len(cleaned_df.PREMON.unique()))
print('Variance : ',cleaned_df.PREMON.var())
print('Null : ',cleaned_df.PREMON.isna().sum())
print('\nThe 99th quantile of pre-monsoon is : ',round(cleaned_df.PREMON.quantile(0.99),2))
print('Number of outliers above 99th quantile is : ',len(cleaned_df[cleaned_df.PREMON > cleaned_df.PREMON.quantile(0.99)]))
plt.title('Box Plot of Pre Monsoon (April to June)')
```

```python
#replacing outliers with lower quantile & uper quantile
cleaned_df.PREMON.clip(cleaned_df.PREMON.quantile(0),cleaned_df.PREMON.quantile(0.99),inplace=True)

cleaned_df.PREMON.plot(kind = 'box')
plt.title('Box Plot of Pre Monsoon (April to June) - After outlier treatment')
```

#### Monsoon (July - September)

```python
cleaned_df.MONSOON.plot(kind = 'box')
print('\nMean : ',cleaned_df.MONSOON.mean())
print('Median : ',cleaned_df.MONSOON.median())
print('Mode : ',cleaned_df.MONSOON.mode()[0])
print('Max : ',cleaned_df.MONSOON.max())
print('Min : ',cleaned_df.MONSOON.min())
print('Unique : ',len(cleaned_df.MONSOON.unique()))
print('Variance : ',cleaned_df.MONSOON.var())
print('Null : ',cleaned_df.MONSOON.isna().sum())
print('\nThe 99th quantile of Monsoon is : ',round(cleaned_df.MONSOON.quantile(0.99),2))
print('The number of values above 99th quantile is : ',len(cleaned_df[cleaned_df.MONSOON > cleaned_df.MONSOON.quantile(0.99)]))
plt.title('Box Plot of Monsoon (Jult to September)')
```

```python
cleaned_df.MONSOON.clip(cleaned_df.MONSOON.quantile(0),cleaned_df.MONSOON.quantile(0.99),inplace=True)
cleaned_df.MONSOON.plot(kind='box')
plt.title('Box Plot of Monsoon (Jult to September) - After outlier treatment')
```

#### Post Monsoon Kharif (October to December)

```python
cleaned_df.POMKH.plot(kind='box')
print('\nMean : ',cleaned_df.POMKH.mean())
print('Median : ',cleaned_df.POMKH.median())
print('Mode : ',cleaned_df.POMKH.mode()[0])
print('Max : ',cleaned_df.POMKH.max())
print('Min : ',cleaned_df.POMKH.min())
print('Unique : ',len(cleaned_df.POMKH.unique()))
print('Variance : ',cleaned_df.POMKH.var())
print('Null : ',cleaned_df.POMKH.isna().sum())
print('\n The 99th quantile of post monsoon kharif is : ',round(cleaned_df.POMKH.quantile(0.99),2))
print('The number of values above 99th quantile is : ',len(cleaned_df[cleaned_df.POMKH > cleaned_df.POMKH.quantile(0.99)]))
plt.title('Box Plot of Post Monsoon Kharif (October to December)')
```

#### Made a copy of the cleaned_df for different seasons

```python
data_POMRB = cleaned_df[['STATE','DISTRICT','YEAR','POMRB']].copy()
data_PREMON = cleaned_df[['STATE','DISTRICT','YEAR','PREMON']].copy()
data_MONSOON = cleaned_df[['STATE','DISTRICT','YEAR','MONSOON']].copy()
data_POMKH = cleaned_df[['STATE','DISTRICT','YEAR','POMKH']].copy()

data_POMRB.head(3)
```

```python
#Renaming the columns with same name
data_POMRB.rename(columns={'POMRB':'Water_level','YEAR':'QUARTER'},inplace = True)
data_PREMON.rename(columns={'PREMON':'Water_level','YEAR':'QUARTER'},inplace = True)
data_MONSOON.rename(columns={'MONSOON':'Water_level','YEAR':'QUARTER'},inplace = True)
data_POMKH.rename(columns={'POMKH':'Water_level','YEAR':'QUARTER'},inplace = True)
```

#### ADDING MONTHS TO QUARTERS
* Post-monsoon Rabi (January to March) -- POMRB
* Pre monsoon (April to June) -- PREMON
* Monsoon (July to September) -- MONSOON
* Post-monsoon Kharif (October to December) -- POMKH

```python
data_POMRB.QUARTER = data_POMRB.QUARTER.apply(lambda x: str(x)+"-01-01")
data_PREMON.QUARTER = data_PREMON.QUARTER.apply(lambda x: str(x)+"-04-01")
data_MONSOON.QUARTER = data_MONSOON.QUARTER.apply(lambda x: str(x)+"-07-01")
data_POMKH.QUARTER = data_POMKH.QUARTER.apply(lambda x: str(x)+"-10-01")
```

#### change to date-time data type

```python
data_POMRB['QUARTER'].dtype
```

```python
data_POMRB.QUARTER = pd.to_datetime(data_POMRB.QUARTER)
data_PREMON.QUARTER = pd.to_datetime(data_PREMON.QUARTER)
data_MONSOON.QUARTER = pd.to_datetime(data_MONSOON.QUARTER)
data_POMKH.QUARTER = pd.to_datetime(data_POMKH.QUARTER)
```

### AutoRegression Model
 A model that uses the dependent relationship between an observation and some number of lagged observations.

```python
train = ldh['Water_level'].loc['1994-01-01':'2015-12-01']
test = ldh['Water_level'].loc['2016-01-01':]
```

```python
from statsmodels.tsa.ar_model import AR
```

```python
model = AR(train)
model_fit = model.fit()
```

```python
model_fit.k_ar
```

```python
model_fit.params
```

```python
predictions = model_fit.predict(start=len(train), end=len(train)+len(test)-1)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(test, predictions)
mse
```

```python
pyplot.plot(test_y)
pyplot.plot(predictions, color='red')
```

#### Walk Forward 

```python
data = train
predict =[]
for t in test:
    model = AR(data)
    model_fit = model.fit()
    y = model_fit.predict(start=len(data), end=len(train)+len(test)-1)
    print(y.values[0])
    predict.append(y.values[0])
    data = np.append(data, t)
    data = pd.Series(data)
    
```

```python
predict
```

```python
mse = mean_squared_error(test.values, predict)
mse
```

```python
pyplot.plot(test.values)
pyplot.plot(predict, color='red')
```
